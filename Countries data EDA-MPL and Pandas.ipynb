#### Q2. What is the distribution of life expectancy across all countries and years?


```python
plt.hist(countries['lifeExpectancy'],
        facecolor="skyblue",
        edgecolor="blue",
        bins=20)
plt.xlabel("Life Expectancy")
plt.ylabel("Frequency")
plt.title("Distribution of Life Expectancy")
plt.show()
```

The distribution of the life expectancy across all countries and years is shown in the histogram above.

#### Q3. How does life expectancy vary over the years for a specific country (e.g., Afghanistan)?
https://www.geeksforgeeks.org/matplotlib-markers-module-in-python/
https://www.geeksforgeeks.org/grids-in-matplotlib/


```python
countryAfghanistan=countries[countries['country']=="Afghanistan"]

fig,ax1=plt.subplots()
fig=plt.figure(figsize=(30,15))
ax1.set_xlabel("Year")
ax1.tick_params(axis="x",
               bottom=False,
               labelbottom=True)
ax1.set_ylabel("Life Expectancy",
              color="black",
              size="large")
ax1.set_title("Life Expectancy Over the Years in Afghanistan")
ax1.tick_params(axis="y",
               labelcolor="black",
               labelsize="large")
ax1.plot(countryAfghanistan['year'],
        countryAfghanistan['lifeExpectancy'],
        color="green",
        marker=".")
ax1.grid(True)
```

#### Q4. What is the relationship between GDP per capita and life expectancy?


```python
plt.scatter(countries['gdpPerCapita'], 
           countries['lifeExpectancy'],
           color = "red")
plt.xlabel("GDP per Capita")
plt.ylabel("Life Expectancy")
plt.title("GDP per Capita vs. Life Expectancy")
plt.grid(True)
plt.show()
```

The relationship shows a correlation where the lower the GDP, the lower the life expectancy and the higher the GDP, the higher the life expectancy.

#### Q5. How has the population changed over time for a specific continent (e.g., Africa)?

** Use groupby for year with populations

```python
africa = countries[countries['continent']=="Africa"]
africa.head()
```

```python
countriesAfrica = africa.groupby(['year'])['population'].sum()
plt.plot(countriesAfrica,
        color = "red",
        marker = ".")
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("Population Over Time in Africa")
plt.grid(True)
plt.show()
```

#### Q6. What is the trend of GDP per capita over the years for a specific continent (e.g., Europe)?


```python
europe = countries[countries['continent']=="Europe"]
europe.head()
```

```python
countriesEurope = europe.groupby(['year'])['gdpPerCapita'].sum()
plt.plot(countriesEurope,
        color = "blue",
        marker = ".")
plt.xlabel("Year")
plt.ylabel("GDP per Capita")
plt.title("GDP per Capita Trend Over Time in Europe")
plt.grid(True)
plt.show()
```
